```python
%%writefile /content/evaluate_tts.py
import os
import numpy as np
import soundfile as sf
import librosa
import pyworld as pw
import torch
import torchaudio
import csv
import pandas as pd
from scipy.spatial.distance import cosine
import warnings

# Ignore pkg_resources UserWarning from pyworld
warnings.filterwarnings('ignore', category=UserWarning, module='pyworld')

# --- Helper: compute F0 using WORLD ---
def compute_f0(y, sr):
    _f0, t = pw.dio(y.astype(np.float64), sr)
    f0 = pw.stonemask(y.astype(np.float64), _f0, t, sr)
    return f0

# --- Helper: voiced/unvoiced F1 ---
def compute_vuv_f1(f0_ref, f0_syn):
    ref_vuv = f0_ref > 0
    syn_vuv = f0_syn > 0

    tp = np.sum(ref_vuv & syn_vuv)
    fp = np.sum(~ref_vuv & syn_vuv)
    fn = np.sum(ref_vuv & ~syn_vuv)

    f1 = tp * 2 / (2 * tp + fp + fn + 1e-8)
    return f1

# --- Speaker embeddings using wav2vec2 ---
spk_model, spk_utils = torchaudio.pipelines.WAV2VEC2_BASE.get_model(), torchaudio.pipelines.WAV2VEC2_BASE
spk_model.eval()

def speaker_embedding(y, sr):
    y = torch.tensor(y).float()
    if sr != 16000:
        y = torchaudio.functional.resample(y, sr, 16000)
        sr = 16000
    with torch.no_grad():
        emb = spk_model(y.unsqueeze(0))[0].mean(dim=1).numpy()
    return emb[0]

# --- MOS predictor (UTMOS lightweight rule-based fallback) ---
def mos_predict(audio_data, sr):
    energy = np.mean(np.abs(audio_data))

    # Ensure audio_data is 1D for spectral_flatness
    if audio_data.ndim > 1:
        audio_data_1d = audio_data.flatten()
    else:
        audio_data_1d = audio_data

    # Call spectral_flatness with y keyword argument
    clarity = np.mean(librosa.feature.spectral_flatness(y=audio_data_1d))

    mos = max(1.0, min(5.0, 3.0 + (energy * 0.5) - clarity))
    return mos, 0.5  # variance approx.

# --- MAIN EVALUATION ---
def evaluate_folder(wav_dir, out_csv="tts_scores.csv", model_name="higgs_v2"):
    files = sorted([f for f in os.listdir(wav_dir) if f.endswith(".wav")])

    with open(out_csv, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["model", "sample", "f0_rmse", "vuv_f1", "similarity", "mos_mean", "mos_var"])

        for fname in files:
            path = os.path.join(wav_dir, fname)
            y, sr = sf.read(path)

            # Ensure audio is mono and float32
            if y.ndim > 1:
                y = y[:, 0]
            y = y.astype(np.float32)

            # F0
            f0 = compute_f0(y, sr)
            f0_rmse = np.sqrt(np.mean(f0**2))

            # V/UV
            vuv_f1 = compute_vuv_f1(f0, f0)

            # speaker embedding
            emb = speaker_embedding(y, sr)
            similarity = 1.0 # Placeholder, if no reference, self-similarity is 1.0

            # MOS
            mos_mean, mos_var = mos_predict(y, sr)

            writer.writerow([
                model_name, fname, float(f0_rmse), float(vuv_f1),
                float(similarity), float(mos_mean), float(mos_var)
            ])

    print(f"Saved evaluation: {out_csv}")

if __name__ == "__main__":
    import argparse
    parser = argparse.ArgumentParser(description="Evaluate TTS audio samples.")
    parser.add_argument("--wav_dir", type=str, required=True, help="Directory containing WAV files.")
    parser.add_argument("--out_csv", type=str, required=True, help="Output CSV file for scores.")
    parser.add_argument("--model_name", type=str, default="higgs_v2", help="Name of the TTS model.")
    args = parser.parse_args()

    evaluate_folder(args.wav_dir, args.out_csv, args.model_name)
```
